```python
# Create a mapping of old values to new shorter values for Q3
short_names_q3 = {
    'United States of America': 'USA',
    'United Kingdom of Great Britain and Northern Ireland': 'UK',
    'Some other long country name': 'Shortened Name'  # Add more mappings as needed
}
# Replace the values in the Q3 column
df['Q3'] = df['Q3'].replace(short_names_q3)

vvc(df, 'Q3', 'x', 'y', 'title', bar_width= 4, bar_spacing= 8,angling=90,label_size=8)
```

```python
# Create a mapping of old values to new shorter values
short_names = {
    'Masters degree': 'Master',
    'Bachelors degree': 'Bachelor',
    'Doctoral degree': 'Doctorate',
    'Some college/university study without earning a bachelors degree': 'Some College',
    'Professional degree': 'Professional',
    'I prefer not to answer': 'No Answer',
    'No formal education past high school': 'High School'
}

# Replace the values in the Q4 column
df['Q4'] = df['Q4'].replace(short_names)
vvc(df, 'Q4', 'x', 'y', 'title')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter out rows where Q24 is 'Unknown'
filtered_df = df[df['Q24'] != 'Unknown']

# Visualize the income (Q24) for all education levels in Q4
plt.figure(figsize=(12, 8))
sns.countplot(data=filtered_df, x='Q24', hue='Q4', order=filtered_df['Q24'].value_counts().index)
plt.title('Income Distribution Across All Education Levels')
plt.xlabel('Income (Q24)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Education Level')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Filter out rows where Q20 is 'Unknown' or NaN
filtered_df = df[df['Q20'].notna() & (df['Q20'] != 'Unknown')]

# Create a bar chart
plt.figure(figsize=(12, 8))
sns.countplot(data=filtered_df, x='Q4', hue='Q20', order=filtered_df['Q4'].value_counts().index)
plt.title('Study Level Preference by Big Companies')
plt.xlabel('Study Level (Q4)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Size')
plt.tight_layout()
plt.show()
```
